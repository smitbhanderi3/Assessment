```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Load Data:

```python
# Load the IRIS dataset
iris = pd.read_csv('iris.csv')
iris
```

### A. Creating and customizing line charts using Matplotlib

```python
# plt.figure(figsize=(8, 4))
sns.lineplot(data=iris, x='Species', y='SepalLengthCm')
plt.title('Line Chart: Sepal Length by Species')
plt.xticks(rotation=45)
plt.show()
```

## B. Visualizing relationships between two or more variables using scatter plots

```python

# Example: Scatter plot of sepal length vs. sepal width
sns.scatterplot(data=iris, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title('Scatter Plot: Sepal Length vs. Sepal Width')
plt.show()
```

## C). Studying distributions of variables using histograms & bar charts to

```python
sns.histplot(data=iris, x='PetalLengthCm', bins=20, kde=True)
plt.title('Histogram: Petal Length Distribution')
plt.show()
```

## D. Visualizing two-dimensional data using heatmaps


```python
corr_matrix = iris.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(data=corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

## E. Displaying images using Matplotlib's plt.imshow 


```python
random_image = np.random.random((10, 10))
plt.imshow(random_image, cmap='viridis')
plt.title('Random Image')
plt.colorbar()
plt.show()
```

## F. Plotting multiple Matplotlib and Seaborn charts in a grid 

```python
sns.set(style='ticks')
sns.pairplot(data=iris, hue='Species', diag_kind='kde', markers=['o', 's', 'D'])
plt.suptitle('Pair Plot: Pairwise Relationships by Species', y=1.02)
plt.show()
```
